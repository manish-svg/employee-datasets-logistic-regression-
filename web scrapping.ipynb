```python
from bs4 import BeautifulSoup
import requests
```

```python
data = requests.get("https://www.flipkart.com/ckf/czl/~cs-3nzcex2cof/pr?sid=ckf,czl&collection-tab-name=Best+Deals+on+TVs&otracker=hp_banner_1_9.bannerX3.BANNER_8BGFNANZESOY&fm=neo%2Fmerchandising&iid=M_c15d16e8-c3ae-4dba-b8c5-384f89883467_9.8BGFNANZESOY&ppt=hp&ppn=homepage&ssid=mj4f3al2ao0000001684466363489").text
soup = BeautifulSoup(data,'lxml')
```

```python
url = []
for i in soup.find_all('a',class_ = 'ge-49M'):
    url.append('https://www.flipkart.com/'+i.get('href'))
```

```python
dicti = {}
```

```python
tv_set = []
rating = []
operating_sys = []
views = []
#delivery_stats = []
price = []
warranty =[]
pixels = []
for urls in url:
    data1 = requests.get (urls).text
    soup1 = BeautifulSoup(data1,'lxml')
    for tv in soup1.find_all('div',class_ = '_4rR01T'):
        tv_set.append(tv.text)
    for ratings in soup1.find_all('div',class_ = 'gUuXy-'):#main class of page items
        # excluding similar products container
        rating1 = ratings.find_all('div',class_ = '_3LWZlK')
        for r in rating1:
            rating.append(r.text)
        # for reviews     
    for reviews in soup1.find_all('span',class_ = '_2_R_DZ'):
        views.append(reviews.text)
        # for operating system
    for op in soup1.find_all('ul',class_ = '_1xgFaf'):
        if 'Tizen' in op.text:
            operating_sys.append('Tizen')
        elif 'Android' in op.text:
            operating_sys.append('Android')
        elif 'Google TV' in op.text:
            operating_sys.append('Google TV')
        else:
            operating_sys.append(None)
        # for price
    for p in soup1.find_all('div',class_ = '_30jeq3 _1_WHN1'):
        price.append(p.text)
        ##### for the warranty column
    for main in soup1.find_all('ul',{"class" : '_1xgFaf'}):
        #getting third sibling
        v = main.find('li',class_ = 'rgWa7D').next_sibling.next_sibling
        if v == None:
            warranty.append(None)
        else :
            warranty.append(v.text)
    ##### for the pixels columns
    for s_main in soup1.find_all('ul',{"class" : '_1xgFaf'}):
        
        #getting second sibling
        k = s_main.find('li',class_ = 'rgWa7D').next_sibling
        if k == None:
            pixels.append(None)
        else:
            k1 = k.text
            if 'Pixels' in k1:
                pixels.append(k1)
            else:
                pixels.append(None)
```

```python
dicti['tv_set'] = tv_set
dicti['rating'] = rating
dicti['operating_sys'] = operating_sys
dicti['views'] = views
#delivery_stats = []
dicti['price'] = price
dicti['warranty'] = warranty
dicti['pixels'] = pixels            
        
        
    
```

# initiating data munging and data cleaning:

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(dicti)
```

```python
data
```

```python
# getting all the brand names
data['Tv_Brands'] = data['tv_set'].apply(lambda x:x.split(' ')[0])
```

```python
# advance splitting 
#import re
data['Tv_Size'] = data['tv_set'].apply(lambda x:x.split("(")[1].split(")")[0])
```

```python
# separating the reviews and ratings coulmns from views
data['Ratings'] = data['views'].apply(lambda x:x.split('&')[0].split(' ')[0])
data['Reiews'] = data['views'].apply(lambda x:x.split('&')[1].split(' ')[0])
```

```python
# excluding none type values for those who do not have warranty
data['Warranty_year'] = data['warranty'].apply(lambda x: x.split()[0].strip() if x != None else (0))
```

```python
data['pixels'] = data['pixels'].apply(lambda x: x.replace('×','x') if x != None else (x))
```

```python
# extracting dimensions from pixels 
data['Dimension_length'] = data['pixels'].apply(lambda x: x.split('x')[0].split()[-1].strip() if x!= None else (np.nan))
data['Dimension_Breadth'] = data['pixels'].apply(lambda x: x.split('x')[1].split()[0].strip() if x!= None else (np.nan))
```

```python
data['Quality'] = data['pixels'].apply(lambda x: x.split(' ')[0:2][0] if x!= None else (np.nan)) +' '+ data['pixels'].apply(lambda x: x.split(' ')[0:2][1] if x!= None else (np.nan))
```

```python
#data['Quality'] = data['pixels'].apply(lambda x: x  if x != None else(np.nan))
```

```python
data['price'] = data['price'].apply(lambda x : x.replace('₹','') if x!= None else (np.nan))
data['price'] = data['price'].apply(lambda x : x.replace(',','') if x!= None else (np.nan))
data['Ratings'] = data['Ratings'].apply(lambda x : x.replace(',','') if x!= None else (np.nan))
data['Reiews'] = data['Reiews'].apply(lambda x : x.replace(',','') if x!= None else (np.nan))
```

```python
data['Dimension_length'] = data['Dimension_length'].astype(str)
data['Dimension_length'] = data['Dimension_length'].apply(lambda x : x.replace('(','') if x!= None else (np.nan))
data['Dimension_Breadth'] = data['Dimension_Breadth'].astype(str)
data['Dimension_Breadth'] = data['Dimension_Breadth'].apply(lambda x : x.replace(')','') if x!= None else (np.nan))
```

```python
data.drop(['tv_set','views','warranty','pixels'],axis = 1, inplace = True)
```

```python
data['Warranty_year'] = data['Warranty_year'].astype(float)
data['Dimension_length'] = data['Dimension_length'].astype(float)
data['Dimension_Breadth'] = data['Dimension_Breadth'].astype(float)
```
